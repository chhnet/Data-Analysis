```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 数据准备
a = np.random.randn(100)
s = pd.Series(a) 

# 用Matplotlib画直方图
#plt.hist(x, bins=10) 函数，其中参数 x 是一维数组，bins 代表直方图中的箱子数量，默认是 10。
plt.hist(s)
plt.show()

# 用Seaborn画直方图
# sns.distplot(x, bins=10, kde=True) 函数。其中参数 x 是一维数组，
# bins 代表直方图中的箱子数量，kde 代表显示核密度估计，默认是 True
sns.distplot(s, kde=False)
plt.show()
sns.distplot(s, kde=True)
plt.show()
```
