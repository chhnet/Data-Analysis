```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 
N = 1000
x = np.random.randn(N)
y = np.random.randn(N)

# Matplotlib
#plt.scatter(x, y, marker=None) xy marker x>o
plt.scatter(x, y,marker='x')
plt.show()

# Seaborn
#xy  data data  DataFrame kind  scatter
df = pd.DataFrame({'x': x, 'y': y})
sns.jointplot(x="x", y="y", data=df, kind='scatter');
plt.show()
```
