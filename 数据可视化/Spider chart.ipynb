```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.font_manager import FontProperties  
'''
首先设置两个数组：labels 和 stats。他们分别保存了这些属性的名称和属性值。
蜘蛛图是一个圆形，需要计算每个坐标的角度，然后对这些数值进行设置。当画完最后一个点后，需要与第一个点进行连线
因为需要计算角度，所以我们要准备 angles 数组；又因为需要设定统计结果的数值，
所以我们要设定 stats 数组。并且需要在原有 angles 和 stats 数组上增加一位，也就是添加数组的第一个元素。
'''

# 数据准备
labels=np.array([u"推进","KDA",u"生存",u"团战",u"发育",u"输出"])
stats=[83, 61, 95, 67, 76, 88]
# 画图数据准备，角度、状态值
angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)
stats=np.concatenate((stats,[stats[0]]))
angles=np.concatenate((angles,[angles[0]]))

# 用Matplotlib画蜘蛛图
fig = plt.figure()  #创建一个空白的 figure 对象
ax = fig.add_subplot(111, polar=True)    #把画板划分成 1 行 1 列
#用 ax.plot 和 ax.fill 进行连线以及给图形上色
ax.plot(angles, stats, 'o-', linewidth=2)
ax.fill(angles, stats, alpha=0.25)

# 设置中文字体
font = FontProperties(fname=r"C:\Windows\Fonts\simhei.ttf", size=14)  
ax.set_thetagrids(angles * 180/np.pi, labels, FontProperties=font)
plt.show()
```
