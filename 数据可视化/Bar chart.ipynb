```python
import matplotlib.pyplot as plt
import seaborn as sns
# 数据准备
x = ['Cat1', 'Cat2', 'Cat3', 'Cat4', 'Cat5']
y = [5, 4, 8, 12, 7]

# 用Matplotlib画条形图
# 参数 x 代表 x 轴的位置序列，height 是 y 轴的数值序列，也就是柱子的高度。
plt.bar(x, y)
plt.show()

# 用Seaborn画条形图
# 使用 sns.barplot(x=None, y=None, data=None) 函数。其中参数 data 为 DataFrame 类型，x、y 是 data 中的变量
sns.barplot(x, y)
plt.show()
```
